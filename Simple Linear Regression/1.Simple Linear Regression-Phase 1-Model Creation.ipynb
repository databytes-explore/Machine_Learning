```python
import pandas as pd
#pandas is one of the libraries in python where the data is stored in the form of table and the dataframe is used for data manipulation and analysis
#library pandas is shortened as pd and refered as pd elsewhere in the code
```

```python
dataset = pd.read_csv("Salary.csv")
dataset
#dataset is created in the form of table by reading the csv file
```

```python
independent = dataset[["YearsExperience"]]
independent
#declaration of independent variable by fetching the x variable from the data set which is nothing but the year of experience
```

```python
dependent = dataset[["Salary"]]
dependent
#declaration of dependent variable by fetching the y variable from the data set which is nothing but the field salary
```

```python
#sklearn is a python library which is used in machine learning . It is used to split training and test data set immporting train_test_split function from sklearn
from sklearn.model_selection import train_test_split
```

```python
#train_test_split takes 4 arguments as input to generator the model which is the independent variable,dependent variable,test size=0.30 means 30% is test data and random state=0 which means how many ever times the function is executed the retain values are same
#and the output is again stored to 4 variables here xtrain= year of experience data for training the model, x_test= years of expericence data for testing the model
#y_train= salary data for training the model and y_test =salary data for testing the model
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
#import Linear regression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
#linear regression class is assigned to a variable
regressor= LinearRegression()
#the training set is now passed to the linear regression function fit which calculates the weight and bias
regressor.fit(x_train,y_train)
```

```python
#fetch the weight value from the model created
weight=regressor.coef_
weight
```

```python
#fetch the bias or the intercept value from the model created
bias=regressor.intercept_
bias
```

```python
#cross check the model by passing the test values
y_pred=regressor.predict(x_test)
x_test
y_pred
```

```python
#calcuate the efficiency of the model which is nothing but rscore rscore close to 1 indicates the model is good
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
#pickle is a library in pythong which is used  to save the model created and deply the same.
import pickle
filename="Finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
result
```
